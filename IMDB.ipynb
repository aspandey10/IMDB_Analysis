```python
#find the unique values from genera
```

```python
my_list=[]
for x in df['Genre']:
    my_list.append(x.split(','))
print(my_list)
```

```python
new_list=[]
for x in my_list:
    for y in x:
        new_list.append(y.split(','))
print(new_list)
```

```python
unique_list=[]
for x in new_list:
    if x not in unique_list:
        unique_list.append(x)
print(unique_list)
```
